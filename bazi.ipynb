```python
import pandas as pd
from google.colab import files

uploaded = files.upload()

# Read the Excel file into a Pandas DataFrame
df = pd.read_excel(io.BytesIO(uploaded['deduped_file14_with_date_columns.xlsx']))

industry_types = df['Industry type'].unique()

industry_elements = {}
industry_elements['metal'] = ['Science', 'Insurance', 'Heavy Metal', 'Finance', 'Investment', 'Automotive', 'Mechanical Engineering']
industry_elements['wood'] = ['Wood Manufacturing', 'Renovation', 'Education', 'Gallery', 'Furniture', 'Horticulture', 'Publishing', 'Healthcare', 'Religious Supplies', 'Religion', 'Paper Manufacturing', 'Plantation', 'Clothing and Textile']
industry_elements['water'] = ['Maritime', 'Tourism', 'Entertainment', 'Legal', 'Design', 'Trade', 'Toy', 'Beverage', 'Aquaculture', 'Logistics', 'Art and Culture', 'Advertising and Marketing', 'Audio', 'Urology, Otorhinolaryngology (ENT)']
industry_elements['fire'] = ['Aviation', 'Beauty', 'Writing', 'Consulting and Auxiliary', 'Data', 'Virtual', 'Catering', 'Metaphysics', 'Energy', 'Aesthetics', 'E-commerce', 'Telecommunications', 'Technology', 'Ophthalmology and Cardiology']
industry_elements['earth'] = ['Funeral', 'Antique', 'Pet', 'Accounting', 'Nursing', 'Real Estate', 'Artifact', 'Cement', 'Construction', 'Agriculture', 'Animal Husbandry', 'Pawnshop', 'Food Processing', 'Infrastructure and Construction']

df['Industry Element'] = 'Other'
for industry_type in industry_types:
    for element, industries in industry_elements.items():
        if industry_type in industries:
            df.loc[df['Industry type'] == industry_type, 'Industry Element'] = element
            break

# Display the updated DataFrame
print(df)

# Download the file to your local machine
df.to_excel('deduped_file13.xlsx', index=False)

# Save the updated DataFrame to a new Excel file
df.to_excel('deduped_file13_with_industry_element.xlsx', index=False)

# Download the file to your local machine
files.download('deduped_file13_with_industry_element.xlsx')
```
